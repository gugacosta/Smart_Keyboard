# Análise de Markov


## Este utiliza a análise de markov simples para gerar recomendações de possíveis palavras para quem está escrevendo um texto.

```python
import string
## Função para ler palavras de um arquivo .txt e armazenar em uma lista
def arquivo_to_list(arquivo_name): 
  """Esta função recebe como entrada um arquivo .txt e retorna uma lista com as palavras contida nele separadas"""

  arq_list = []
  fin = open(arquivo_name)   ##

  for line in fin:
    line = line.strip(string.whitespace)
    line = line.split()
    
    if len(line)>1:
      for word in line:
        word = word.lower()         
        arq_list.append(word) 

    else:
      if line==[]:
        line = '\n'
      line = str(line)   
      arq_list.append(line)

  return arq_list

def markov_analyses(list_words):
  """Esta função recebe como entrada uma lista de palavras em seuquencia e retorna um dict os possíveis sufixos de acordo com a frequência"""

  markov_dict = {}
  for x in range(0,len(list_words)-2):
    if list_words[x] not in markov_dict:
        markov_dict[list_words[x]] = [[1,list_words[x+1]]]

    else:
        flag = False
        for sufix in markov_dict[list_words[x]]:
            if sufix[1] == list_words[x+1]:
                sufix[0] = sufix[0] + 1
                flag = True
            
        if flag == False:
            markov_dict[list_words[x]].append([1,list_words[x+1]])             

        markov_dict[list_words[x]].sort(reverse=True)
  
  return markov_dict



teste = arquivo_to_list("emma.txt")
markov_dict = markov_analyses(teste)
ler_palavra = ""
while ler_palavra != 'end':
   ler_palavra = str(input("Digite uma palavra e vamos sugerir os sufixos mais prováveis "))
   print(markov_dict[ler_palavra][:4])
```
